Aula de Programação e Algoritmos
* calcular a área de uma sala

# Entrada

   Medir o comprimento da sala e medir a largura da sala

# Processamento

   Fazer a multiplicação do comprimento pela largura

# Saída
  Mostrar o resultado


```python
# Entrada

comprimento = 7 # metros
largura = 4 # metros
```

```python
# Processamento

area = comprimento * largura
```

```python
print (f"o resultado da área da sala é: {area} metros quadrados")
```

```python
#Calcular o quadrado

#Entrada

lado = 2 # metros

#Processamento

resultado = lado * lado

#Saída

print (f"o resultado da área do quadrado é: {resultado} metros quadrados")
```

* Escrever um algoritmo e o programa que receba 4 notas do aluno e calcula a média dessas notas e mostre a média.


# Entrada
 Escrever cada uma das 4 notas do aluno

# Processamento
 Somar todas as notas do aluno e dividir pelo número de notas

# Saída
 Resultado da média do aluno

```python
# Entrada

numero_1 = 9
numero_2 = 10
numero_3 = 8
numero_4 = 10
```

```python
# Processamento

igual = (numero_1 + numero_2 + numero_3 + numero_4)/ 4
```

```python
# Saída
print (f"a média do aluno é:{igual}")
```
