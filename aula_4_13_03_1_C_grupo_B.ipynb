* **ESCREVER UM PROGRAMA QUE RECEBE A TEMPERATURA EM GRAUS CELSIUS E CONVERTE PARA FAHRENHEIT**

# Entrada

graus celsius

# Processamento

F= C * (9/5) + 32

# Saída

mostrar graus em Fahrenheit





```python
# Escrever um programa que recebe em graus celsius e transforma em fahrenheit

# Entrada

Celsius = float (input ("digite a temperatura em graus celsius:"))

# Processamento

Fahrenheit = Celsius * (9/5) + 32

# Saída

print (f"A temperatura é de {Fahrenheit}°F")

```

```python
# Escreva um programa que soma dois números

# Entrada

a = float (input ("digite o primeiro número: "))
b = float (input ("digite o segundo número: "))

# Processamento

soma = a + b
subtraçao = a - b
multiplicaçao = a * b
divisao = a / b
div_int = a // b # usar barras duplas "//" para aproximação/divisões inteiras
resto = a % b #o símbolo de "%" é igual ao resto da divisão

# Saída


print ("--------------------CALCULADORA SIMPLES------------------- |")
print ("                                                           |")
print ("---------------------------soma--------------------------- |")
print ("                                                           |")
print (f"a soma de {a} mais {b} é igual: {soma} ")
print ("                                                           |")
print ("-------------------------subtração------------------------ |")
print ("                                                           |")
print (f"a subtração de {a} menos {b} é igual: {subtraçao} ")
print ("                                                           |")
print ("-----------------------multiplicação---------------------- |")
print ("                                                           |")
print (f"a multiplicação de {a} vezes {b} é igual: {multiplicaçao}")
print ("                                                           |")
print ("--------------------------divisão------------------------- |")
print ("                                                           |")
print (f"a divisão de {a} dividido por {b} é igual: {divisao}")
print ("                                                           |")
print ("----------------------divisão aproximada------------------ |")
print ("                                                           |")
print (f"a divisão aproximada de {a} dividido por {b} é igual: {div_int}")
print ("                                                           |")
print ("-----------------------------resto------------------------ |")
print ("                                                           |")
print (f"o resto de {a} dividido por {b} é igual: {resto}")
print ("                                                           |")
print ("---------------------------------------------------------- |")
```
