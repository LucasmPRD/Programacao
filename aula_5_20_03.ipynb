```python
#  1. Soma simples

#Faça um programa que:
	#•	Leia dois números inteiros
	#•	Mostre a soma deles


# Entrada
print ("----------------------------SOMA----------------------------")
print ( )
numero_1 = float (input ("digite o primeiro número: "))
numero_2 = float (input ("digite o segundo número: "))
print ( )

# Processamento

soma = numero_1 + numero_2

# Saída

print (f"o resultado da soma de {numero_1} mais {numero_2} é:{soma}")
```

```python
# 2. Subtração

#Crie um algoritmo que:
	#•	Leia dois números
	#•	Mostre o resultado da subtração do primeiro pelo segundo

# Entrada

print ("------------------SUBTRAÇÃO------------------")
print ( )
numero_1 = float (input ("digite o primeiro número: "))
numero_2 = float (input ("digite o segundo número: "))
print ( )

# Processamento

subtracao = numero_1 - numero_2

# Saída

print (f"O resultado da subtração de {numero_1} menos {numero_2} é: {subtracao}")
```

```python
# 3. Multiplicação

#Desenvolva um programa que:
	#•	Leia dois números
	#•	Mostre o resultado da multiplicação

# Entrada
print ("--------------MULTIPLICAÇÃO---------------")
print ( )
numero_1 = float (input ("digite o primeiro número:"))
numero_2 = float (input ("digite o segundo número:"))
print ( )

# Processamento

multiplicacao = numero_1 * numero_2

# Saída

print (f"o resultado da multiplicação de {numero_1} vezes {numero_2} é: {multiplicacao}")
```

```python
#4. Divisão

#Faça um programa que:
	#•	Leia dois números
	#•	Mostre o resultado da divisão

# Entrada

print ("--------------------DIVISÃO----------------")
print ( )
numero_1 = float (input ("digite o primeiro número: "))
numero_2 = float (input ("digite o segundo número: "))
print ( )

# Processamento

divisao = numero_1 / numero_2

# Saída

print (f"O resultado da divisão de {numero_1} por {numero_2} é: {divisao}")
```

```python
#5. Média de 3 números

#Crie um algoritmo que:
	#•	Leia três números
	#•	Calcule e mostre a média

# Entrada

print ("-------------------MÉDIA DE TRÊS NÚMEROS-------------------")
print ( )
numero_1 = float (input ("digite o primeiro número: "))
numero_2 = float (input ("digite o segundo número: "))
numero_3 = float (input ("digite o terceiro número: "))
print ( )

# Processamento

média = (numero_1 + numero_2 + numero_3) / 3

# Saída

print (f"O resultado da média de {numero_1}, {numero_2} e {numero_3} números é: {média}")
```

```python
#6. Divisão inteira e resto

#Faça um programa que:
	#•	Leia um número inteiro
	#•	Mostre:
	#•	O dobro
	#•	A metade
	#•	O resto da divisão por 3

# Entrada

a = int (input ("digite o número: "))
print ( )

# Processamento

dobro = 2 * a
metade = a / 2
resto = a % 3

# Saída

print (f"o dobro do número {a} é igual: {dobro}")
print (f"a metade do número {a} é igual: {metade}")
print (f"o resto do número {a} é igual: {resto}")

```

```python
#7. Conversão de temperatura

#Desenvolva um programa que:
	#•	Leia uma temperatura em Celsius
	#•	Converta para Fahrenheit


# Entrada

print ("----------------------------TERMÔMETRO CONVERSOR--------------------------------")
print ( )
Celsius = float (input ("digite a temperatura em graus celsius:"))
print ( )

# Processamento

Fahrenheit = Celsius * (9/5) + 32

# Saída

print ("----------------------------COVERSÃO EM FAHRENHEIT--------------------------------")
print ( )
print (f"A temperatura em Fahrenheit é de {Fahrenheit}°F")
```

```python
#8. Cálculo de salário

#Faça um algoritmo que:
	#•	Leia o salário de um funcionário
	#•	Calcule um aumento de 10%
	#•	Mostre o novo salário

# Entrada

print ("-----------------AUMENTO DE SALÁRIO---------------")
salário = float (input ("digite o valor do salário:"))
print ( )
# Processamento

aumento = salário + (salário * 10/100)

# Saída

print (f"O sálario com o aumento de 10% é igual: {aumento}")
```

```python
# 9. Área de um retângulo

#Crie um programa que:
	#•	Leia a base e a altura
	#•	Calcule a área

# Entrada

print ("----------------------ÁREA DO RETÂNGULO-------------------")
print ( )
comprimento = float (input ("digite o comprimento do retângulo em metros:"))
largura = float (input ("digite a largura do retângulo em metros:"))
print ( )

# Processamento

área = comprimento * largura

# Saída

print (f"a área do retângulo é de: {área} metros quadrados")
```

```python
#10. Consumo de combustível

#Faça um programa que:
	#•	Leia a distância percorrida (km)
	#•	Leia o combustível gasto (litros)
	#•	Calcule o consumo médio (km/l)

# Entrada

print ("--------------------KM/L---------------------")
print ( )
km = float (input ("digite a quantidade de quilômetros percorridos:"))
l = float (input ("digite a quantidade de litros gastos:"))
print ( )

# Processamento

consumo = km / l

# Saída

print (f"o consumo médio foi de: {consumo} km/l ")
```
